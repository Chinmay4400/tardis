# Plotting Spectrum Interactively ##

In this notebook, we will plot the spectrum from quickstart notebook interactively

```python
from tardis import run_tardis
from tardis.io.atom_data.util import download_atom_data
```

```python
# the data is automatically downloaded
download_atom_data('kurucz_cd23_chianti_H_He')
```

```python
#TARDIS now uses the data in the data repo
!curl -O https://raw.githubusercontent.com/tardis-sn/tardis/master/docs/models/examples/tardis_example.yml
sim = run_tardis('tardis_example.yml')
```

```python
import bqplot as bq
from bqplot import pyplot as plt, LinearScale, Axis

x_scale = LinearScale()
y_scale = LinearScale()

#Defining x and y axes
x_axis = Axis(label='Wavelength [Å]',scale = x_scale)
y_axis = Axis(label='Luminosity [erg/s/Å]', scale = y_scale, orientation='vertical')

# Defining various spectrums
spectrum = sim.runner.spectrum
spectrum_virtual = sim.runner.spectrum_virtual
spectrum_integrated = sim.runner.spectrum_integrated

# Saving plots
plt1 = plt.plot(spectrum.wavelength, spectrum.luminosity_density_lambda,
                'g-',scales={'x': x_scale, 'y': y_scale},labels=['luminosity density'],display_legend=True)
plt2 = plt.plot(spectrum.wavelength, spectrum_virtual.luminosity_density_lambda, 
                'r-',scales={'x': x_scale, 'y': y_scale},labels=['virtual luminosity density'],display_legend=True)
plt3 = plt.plot(spectrum.wavelength, spectrum_integrated.luminosity_density_lambda, 
                'b-',scales={'x': x_scale, 'y': y_scale},labels=['formal integral'],display_legend=True)
```

**Defining ipywidgets**

```python
from ipywidgets import *
from IPython.display import display, Javascript

luminousity_density = widgets.Checkbox(
    value=True,
    description='Luminousity Density',
    disabled=False,
    indent=True
)

v_luminousity_density = widgets.Checkbox(
    value=True,
    description='Virtual Luminous Density',
    disabled=False,
    indent=True
)

formal_integral = widgets.Checkbox(
    value=True,
    description='Formal Integral',
    disabled=False,
    indent=True
)

show = widgets.Button(
    description='Show',
    disabled=False,
    button_style='',
    tooltip='Show Plots',
    icon='check'
)

buttons_box = HBox(
    children=(luminousity_density,v_luminousity_density,formal_integral,show),
   layout=Layout(margin='left')
)
```

```python
fig = bq.Figure(
    marks=[plt1,plt2,plt3],
    layout=Layout(width='800px', height='500px', margin='0 auto'),
    axes=[x_axis, y_axis],
    legend_style = {'stroke-width': 0},
    title = "Spectrum"
)
```

**Show plots upon click event**

```python
def show_plots(b):
    plots=[]
    if(luminousity_density.value):
        plots.append(plt1)
    if(v_luminousity_density.value):
        plots.append(plt2)
    if(formal_integral.value):
        plots.append(plt3)   
    fig.marks = plots
    return fig
show.on_click(show_plots)
```

**Plot spectrum with chosen plots**

```python
plot_spectrum = VBox(
    children=(buttons_box, fig,),
    layout=Layout(margin='50px 0')
)
plot_spectrum
```
